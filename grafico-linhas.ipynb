```python
# importando libs
import random
import matplotlib.pyplot as plt

# gerando lista de números aleatórios
dados = [random.randint(1, 100) for _ in range(10)]

# exibir lista gerada
print("Lista de dados: ", dados)

# plotar os dados
plt.plot(dados, marker="o") # marcador em cada ponto para melhor visualização

plt.title("Gráfico de números aleatórios")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.grid(True)
plt.show()
```
